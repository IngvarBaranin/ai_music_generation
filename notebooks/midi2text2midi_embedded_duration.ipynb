```python
from music21 import stream
from music21 import converter, corpus, instrument, midi, note, chord, pitch
from music21 import *
import os 
```

```python
def open_midi(midi_path):
    mf = converter.parse(midi_path)
    return mf
```

```python
# Access midifile with Parts merged together with correct offsets

previousElementOffset = 0.0
offsetChanged = False

tempoRetrieved = False
timeSigRetrieved = False

tokens = []

tokens.append("START")
#print("START")

for element in mf.flat.elements:
    #print(type(element))
    
    currentElementOffset = element.offset
    
    if (currentElementOffset != previousElementOffset and (isinstance(element, note.Note) or isinstance(element, chord.Chord))):
        offsetChanged = True
        tokens.append("wait:" + str(currentElementOffset - previousElementOffset))
        #print("wait:" + str(currentElementOffset - previousElementOffset))
        
    if (isinstance(element, tempo.MetronomeMark) and not tempoRetrieved):
        tempoRetrieved = True
        tokens.append("tempo:" + str(element.number))
        #print("tempo" + str(element.number))
        
    if (isinstance(element, meter.TimeSignature) and not timeSigRetrieved):
        timeSigRetrieved = True
        tokens.append("timesig:" + str(element.ratioString))
        #print("timesig:" + str(element.ratioString))
    
    if (isinstance(element, note.Note)): # This is a note event
        tokens.append("note:" + str(element.pitch) + ":v" + str(element.volume.velocity) + ":d" + str(element.duration.quarterLength))
        #print("note:" + str(element.pitch) + ":v" + str(element.volume.velocity) + ":d" + str(element.duration.quarterLength))
        
    if (isinstance(element, chord.Chord)): # This is a chord event
        for chordnote in element: 
            tokens.append("note:" + str(chordnote.pitch) + ":v" + str(chordnote.volume.velocity) + ":d" + str(chordnote.duration.quarterLength))
            #print("note:" + str(chordnote.pitch) + ":v" + str(chordnote.volume.velocity) + ":d" + str(chordnote.duration.quarterLength))
            
    if (offsetChanged):
        previousElementOffset = currentElementOffset
        offsetChanged = False
        
tokens.append("END")
#print("END")

print(tokens)
```

```python
#mf.show("text")
```

```python
s = stream.Stream()

currentOffset = 0

for token in tokens:
    
    splitToken = token.split(":")
    
    if token.startswith("tempo"):
        s.append(tempo.MetronomeMark(number=float(splitToken[1])))
        
    if token.startswith("timesig"):
        s.append(meter.TimeSignature(splitToken[1]))
        
    if token.startswith("note"):
        noteDuration = splitToken[3][1:]
        if "/" in noteDuration:
            numerator = float(noteDuration.split("/")[0])
            denominator = float(noteDuration.split("/")[1])
            noteDuration = numerator / denominator
        
        newNote = note.Note(nameWithOctave=splitToken[1], 
                           velocity=int(splitToken[2][1:]), 
                           quarterLength=float(noteDuration))
        s.insert(currentOffset, newNote)
        
    if token.startswith("wait"):
        currentOffset += float(splitToken[1]) 
        
s.flat.elements
```
